# Problem 1: Mobile Recharge Validation System
## Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be â‰¥ â‚¹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid

## Real-Time Use
Telecom recharge apps
Payment gateways


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    # Check minimum amount
    if amount < 50:
        print("Recharge amount must be at least â‚¹50.")
        return False
    
    # Check valid plans
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False
    
    return True


# Main Program
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: â‚¹"))
        
        if validate_recharge(recharge_amount):
            print("Recharge Successful!")
            break   # Exit loop if valid
        else:
            print("ðŸ”„ Please try again.\n")
    
    except ValueError:
        print("Please enter a valid number.\n")
```

# Problem 2: Inventory Reorder Alert System
## Problem Statement
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise

Real-Time Use
Warehouse systems
Retail management



```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} â†’ Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} â†’ Stock OK (Stock: {stock})")


# Sample Inventory Dictionary
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 18,
    "Salt": 5,
    "Wheat": 30
}

# Function Call
check_inventory(inventory)
```


# Problem 3: Student Result Processing System
## Problem Statement
Create a function that:
Accepts a list of student marks
Calculates average marks
Prints Pass if average â‰¥ 50 else Fail

Real-Time Use
Online exam systems
Evaluation portals



```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total = total + mark
    
    average = total / len(marks)
    
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    
    print("Average Marks:", average)
    return result


# Example Usage
student_marks = [60, 45, 70, 50, 40]

final_result = process_result(student_marks)
print("Result:", final_result)
```

# Problem 4: Cab Fare Estimator with Retry Option
## Problem Statement
Create a cab fare calculator:
Base fare = â‚¹50
â‚¹12 per km
Peak hour â†’ 25% extra
Allow retry using while loop

Real-Time Use
Ride-hailing apps
Transport systems



```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    # Apply peak hour charges
    if peak.lower() == "yes":
        fare = fare + (0.25 * fare)
    
    return fare


# Main Program with Retry Option
while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")
        
        total_fare = calculate_fare(km, peak_hour)
        
        print("Total Fare: â‚¹", total_fare)
        
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator ðŸš–")
            break
    
    except ValueError:
        print("âŒ Please enter valid input.\n")
```

# Problem 5: Employee Attendance Eligibility Checker
## Problem Statement
Create a function that:
Accepts attendance list ("P", "A")
Calculates attendance percentage
Returns Eligible if â‰¥ 75% else Not Eligible

Real-Time Use
HR payroll systems
Internship tracking


```python
def check_attendance(attendance_list):
    present_days = 0
    
    # Count present days using loop
    for day in attendance_list:
        if day == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    
    # Avoid division by zero
    if total_days == 0:
        return "No attendance data"
    
    percentage = (present_days / total_days) * 100
    
    print("Attendance Percentage:", percentage, "%")
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example Usage
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

result = check_attendance(attendance)
print("Status:", result)
```

# Problem 6: Password Strength Checker
## Problem Statement
Create a function to check password strength.
Rules:
Minimum length = 8
Must contain at least one digit
Must contain at least one special character (@#$)

Real-Time Use
Login systems
Security validation


```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    # Final validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"


# Example Usage
user_password = input("Enter password: ")
result = check_password(user_password)
print(result)
```
