## Problem 1: Mobile Recharge Validation System
# Problem Statement
Create a function to validate a mobile recharge request.
Rules:
Recharge amount must be â‰¥ â‚¹50
Valid plans: 199, 299, 399, 599
User should be allowed to retry if invalid

## Real-Time Use

Payment gateways


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    # Check minimum amount
    if amount < 50:
        print("Recharge amount must be at least â‚¹50.")
        return False
    
    # Check valid plans
    if amount not in valid_plans:
        print("Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False
    
    return True


# Main Program
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: â‚¹"))
        
        if validate_recharge(recharge_amount):
            print("Recharge Successful!")
            break   # Exit loop if valid
        else:
            print("ðŸ”„ Please try again.\n")
    
    except ValueError:
        print("Please enter a valid number.\n")
```
