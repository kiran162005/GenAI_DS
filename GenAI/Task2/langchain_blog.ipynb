## 1. Introduction to LangChain

LangChain is a framework used to build applications powered by Large Language Models (LLMs). It helps connect LLMs with tools, memory, and workflows.

### Why LangChain?
- Simplifies LLM development
- Enables chaining of components
- Supports memory and agents

### Problems it Solves
- Prompt management
- Workflow orchestration
- Tool usage

```python
# Load HuggingFace Model
import warnings
warnings.filterwarnings('ignore')

!pip install langchain-community

# Load HuggingFace Model
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

# Load free model
pipe = pipeline("text-generation", model="gpt2")

llm = HuggingFacePipeline(pipeline=pipe)

print("Model loaded successfully")
```

```python
response = llm.invoke("What is LangChain?")
print(response)
```

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms"
)

print(template.format(topic="LangChain"))
```

```python
chain = template | llm

result = chain.invoke({"topic": "Artificial Intelligence"})
print(result)
```

```python
# Simulated Memory

# Create empty memory
chat_history = []

# Store conversations
chat_history.append({"input": "Hi", "output": "Hello"})
chat_history.append({"input": "What is AI?", "output": "Artificial Intelligence"})
chat_history.append({"input": "What is LangChain?", "output": "A framework for building LLM applications"})

# Display memory
print("Conversation History:\n")

for i, chat in enumerate(chat_history, 1):
    print(f"Conversation {i}:")
    print("User:", chat["input"])
    print("Bot :", chat["output"])
    print("-" * 30)
```
