## 1. Introduction to LangChain

**LangChain** is an open-source Python framework for building applications powered by Large Language Models (LLMs).

| Problem with raw LLMs | LangChain Solution |
|---|---|
| No memory across turns | Memory modules |
| Can't access external data | Document Loaders + Vector Stores (RAG) |
| Can't take real-world actions | Agents + Tools |
| Brittle hardcoded prompts | Prompt Templates |
| Hard to chain multiple calls | LCEL Chains (`|` operator) |

## 2. Core Components

### 2.1 LLMs and Chat Models

LangChain wraps all LLM providers behind a unified interface.  
We use `google/flan-t5-large` via HuggingFace Inference API — free and capable.

````python
from langchain_huggingface import HuggingFaceEndpoint

# Initialize LLM — HuggingFace Inference API (free)
llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=256,
    temperature=0.7,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

# Basic LLM call
response = llm.invoke("What is a transformer model? Answer in one sentence.")
print("Response:", response)
````

### 2.2 Prompts and Prompt Templates

A `PromptTemplate` separates prompt structure from content — making prompts reusable and testable.  
**No API key needed for this cell** — pure Python.

````python
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate

# Simple string template with named variables
basic_template = PromptTemplate.from_template(
    "Explain {concept} to a {audience} in {num_sentences} sentences."
)

formatted = basic_template.format(
    concept="recursion",
    audience="10-year-old",
    num_sentences="2"
)
print("Formatted prompt:", formatted)

# Chat prompt template (system + user messages)
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in {domain}. Be precise."),
    ("human", "Question: {question}")
])

messages = chat_template.format_messages(
    domain="machine learning",
    question="What is gradient descent?"
)
print("\nChat messages:")
for msg in messages:
    print(f"  [{msg.type}]: {msg.content}")
````

### 2.3 Chains — LCEL (LangChain Expression Language)

LCEL connects `prompt → LLM → parser` using the `|` pipe operator.  
This is the modern LangChain 0.3.x approach — composable, streamable, easy to test.

````python
from langchain_huggingface import HuggingFaceEndpoint
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=200,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

# Define the prompt
prompt = PromptTemplate.from_template(
    "You are a Python expert. Write a Python function that {task}. Only code, no explanation."
)

# Build chain using the LCEL pipe | operator
chain = prompt | llm | StrOutputParser()

# Invoke the chain
result = chain.invoke({"task": "reverses a list without using the reverse() method"})
print("Generated code:")
print(result)
````

### 2.4 Memory

Memory modules persist conversation history across LLM calls.  
We demonstrate `ConversationBufferMemory` — injecting history into the prompt context.

````python
from langchain.memory import ConversationBufferMemory
from langchain_huggingface import HuggingFaceEndpoint
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Create memory object
memory = ConversationBufferMemory(memory_key="history", return_messages=False)

# Turn 1: save context manually
memory.save_context(
    {"input": "My name is Rahul. I am learning LangChain."},
    {"output": "Nice to meet you Rahul! LangChain is a great framework for AI apps."}
)

# Inspect memory
history = memory.load_memory_variables({})["history"]
print("--- Stored conversation history ---")
print(history)

# Turn 2: inject history into prompt for contextual answer
llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=100,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

prompt_with_history = PromptTemplate.from_template(
    "Previous conversation:\n{history}\n\nUser: {input}\nAssistant:"
)

chain = prompt_with_history | llm | StrOutputParser()

response = chain.invoke({
    "history": history,
    "input": "What is my name and what am I learning?"
})

print("\n--- Model response (with memory context) ---")
print(response)

# Save Turn 2 to memory
memory.save_context(
    {"input": "What is my name and what am I learning?"},
    {"output": response}
)
print(f"\nTotal messages stored in memory: {len(memory.chat_memory.messages)}")
````

### 2.5 Agents

Agents use an LLM as a reasoning engine to decide which tools to call — following the **ReAct** loop: Think → Act → Observe → repeat until done.

````python
from langchain_huggingface import HuggingFaceEndpoint
from langchain.tools import tool
from langchain.agents import AgentExecutor, create_react_agent
from langchain_core.prompts import PromptTemplate

# Use a stronger instruction-following model for agent reasoning
agent_llm = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    max_new_tokens=512,
    temperature=0.1,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

# Define tools using the @tool decorator
@tool
def multiply(a: float, b: float) -> float:
    """Multiplies two numbers. Use for any multiplication task."""
    return a * b

@tool
def add_numbers(a: float, b: float) -> float:
    """Adds two numbers together."""
    return a + b

@tool
def get_string_length(text: str) -> int:
    """Returns the number of characters in a string."""
    return len(text)

tools = [multiply, add_numbers, get_string_length]

# ReAct prompt template — the LLM follows Thought/Action/Observation format
react_prompt = PromptTemplate.from_template("""
Answer the question using the tools available.

Tools:
{tools}

Format:
Question: the question
Thought: reason about what to do
Action: tool name (one of [{tool_names}])
Action Input: input to the tool
Observation: tool result
Thought: I now know the answer
Final Answer: the final answer

Question: {input}
Thought: {agent_scratchpad}
""")

# Create agent and executor
agent = create_react_agent(llm=agent_llm, tools=tools, prompt=react_prompt)
agent_executor = AgentExecutor(
    agent=agent,
    tools=tools,
    verbose=True,
    max_iterations=4,
    handle_parsing_errors=True
)

result = agent_executor.invoke({"input": "What is 45.5 multiplied by 8.2?"})
print("\nFinal Answer:", result["output"])
````

### 2.6 Tools

Tools are functions with a `name`, `description` (LLM reads this to decide when to call them), and an optional `args_schema`.  
**No API key needed** — this demonstrates the tool API directly.

````python
from langchain.tools import tool, StructuredTool
from pydantic import BaseModel

# Simple decorator-based tool
@tool
def get_word_count(text: str) -> int:
    """Returns the number of words in a text string."""
    return len(text.split())

# Structured tool with Pydantic input validation
class WeatherInput(BaseModel):
    city: str
    unit: str = "celsius"

def fetch_weather(city: str, unit: str = "celsius") -> str:
    """Mock weather function — replace with real API (e.g. OpenWeatherMap) in production."""
    temps = {"delhi": 34, "mumbai": 30, "bangalore": 24, "hyderabad": 32, "chennai": 35}
    temp = temps.get(city.lower(), 28)
    return f"The temperature in {city} is {temp} degrees {unit}."

weather_tool = StructuredTool.from_function(
    func=fetch_weather,
    name="get_weather",
    description="Fetches current temperature for a city. Use when user asks about weather.",
    args_schema=WeatherInput
)

# Inspect tool properties (the LLM reads name + description to decide when to call)
print("Tool name:       ", get_word_count.name)
print("Tool description:", get_word_count.description)
print("Direct call:     ", get_word_count.run("Hello world from LangChain"), "words")

print("\nWeather tool name:", weather_tool.name)
print("Delhi weather:    ", weather_tool.run({"city": "Delhi", "unit": "celsius"}))
print("Hyderabad weather:", weather_tool.run({"city": "Hyderabad", "unit": "celsius"}))
````

### 2.8 Indexes and Vector Stores

Chunks are embedded into dense vectors using `sentence-transformers` (runs **locally, no API key**) and stored in FAISS for fast semantic search.

````python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document

# Local embeddings — downloads ~90MB model on first run, then cached
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

# Knowledge base
raw_texts = [
    "LangChain was created by Harrison Chase in October 2022.",
    "LangChain supports both Python and JavaScript.",
    "LCEL uses the pipe | operator for composing LangChain chains.",
    "Agents in LangChain follow the ReAct (Reasoning + Acting) framework.",
    "FAISS is a vector store library developed by Meta AI Research.",
    "LangSmith is the observability platform for LangChain apps.",
    "LangGraph enables stateful multi-actor agentic workflows."
]

docs = [Document(page_content=t, metadata={"id": i}) for i, t in enumerate(raw_texts)]

# Split into chunks (important for large PDFs)
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
chunks = splitter.split_documents(docs)
print(f"Total chunks: {len(chunks)}")

# Build FAISS vector store locally
vector_store = FAISS.from_documents(chunks, embeddings)

# Semantic similarity search
query = "Who created LangChain and when?"
results = vector_store.similarity_search(query, k=2)

print(f"\nQuery: '{query}'")
print("Top results:")
for i, doc in enumerate(results):
    print(f"  [{i+1}] {doc.page_content}")
````

## 3. Architecture Diagram

```
                        User Input
                            │
                            ▼
                    Prompt Template
               (injects system context)
                            │
                            ▼
                  LLM / Chat Model
          (HuggingFace, OpenAI, Ollama...)
                            │
                            ▼
                          Chain
                (LCEL pipe | operator)
                    ╱           ╲
                   ╱             ╲
             Tools / Agents     Memory
          (search, APIs, code)  (history)
                   ╲             ╱
                    ╲           ╱
                        Output

Document Loaders → Text Splitter → Embeddings → Vector Store
                                                       │
                                                  Retriever ──► Chain (injected as context)
                                                  [This is the RAG pattern]
```

**Flow step-by-step:**
1. User query enters the system
2. PromptTemplate formats it with instructions and optional few-shot examples
3. LLM processes the formatted prompt and generates a response
4. Chain orchestrates the call and routes the output
5. If dynamic action is needed → Agent selects and calls a Tool
6. Memory injects prior conversation history for continuity
7. Retriever injects relevant document chunks (RAG) as context
8. Final answer is returned to the user

## 4. Hands-on: Full RAG Pipeline

End-to-end: local embeddings + FAISS + HuggingFace LLM in one runnable pipeline.

````python
from langchain_huggingface import HuggingFaceEndpoint, HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain.schema import Document

# --- Knowledge base ---
texts = [
    "LangChain was created by Harrison Chase in October 2022.",
    "LangChain supports Python and JavaScript.",
    "LCEL uses the pipe | operator for composing chains.",
    "Agents use the ReAct (Reasoning + Acting) framework.",
    "FAISS is a vector similarity search library by Meta AI.",
    "LangSmith provides observability and tracing for LangChain apps.",
    "LangGraph enables stateful multi-actor agentic workflows."
]

# --- Local embeddings + FAISS (no API key needed for this part) ---
docs = [Document(page_content=t) for t in texts]
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = FAISS.from_documents(docs, embeddings)
retriever = vector_store.as_retriever(search_kwargs={"k": 2})

# --- HuggingFace LLM ---
llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=200,
    temperature=0.1,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

# --- RAG prompt ---
rag_prompt = PromptTemplate.from_template(
    """Answer using ONLY the context below.
If the answer is not in the context, say 'I do not know.'

Context:
{context}

Question: {question}
Answer:"""
)

def format_docs(docs):
    """Concatenate retrieved chunks into a single context string."""
    return "\n".join(doc.page_content for doc in docs)

# --- RAG chain using LCEL ---
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | rag_prompt
    | llm
    | StrOutputParser()
)

# --- Test queries ---
questions = [
    "Who created LangChain and when?",
    "What does LCEL stand for and what does it use?",
    "What is LangSmith used for?"
]

for q in questions:
    answer = rag_chain.invoke(q)
    print(f"Q: {q}")
    print(f"A: {answer}\n")
````

## 5. Real-World Use Cases

### Use Case 1 — Customer Support Chatbot

**Problem**: E-commerce company needs a bot that answers questions about orders and policies using an FAQ knowledge base.

**Components**: `Document` + `HuggingFaceEmbeddings` + `FAISS` + RAG Chain

````python
from langchain_huggingface import HuggingFaceEndpoint, HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain.schema import Document

# FAQ knowledge base
faq_docs = [
    Document(page_content="Refund policy: Customers may return items within 30 days for a full refund."),
    Document(page_content="Shipping: Standard delivery is 3-5 business days. Express is 1-2 days."),
    Document(page_content="Order tracking: Visit our website and enter your order ID in the tracking portal."),
    Document(page_content="Damaged items: Contact support within 48 hours with a photo. Replacement ships in 24 hours."),
    Document(page_content="Payment: We accept credit card, debit card, UPI, and net banking."),
]

# Build retriever
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vs = FAISS.from_documents(faq_docs, embeddings)
retriever = vs.as_retriever(search_kwargs={"k": 2})

llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=150,
    temperature=0.3,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

support_prompt = PromptTemplate.from_template(
    """You are a friendly customer support agent.
Use the FAQ below to answer. Be concise.

FAQ:
{context}

Customer: {question}
Agent:"""
)

def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)

support_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | support_prompt
    | llm
    | StrOutputParser()
)

queries = [
    "What is your return policy?",
    "My item arrived damaged. What should I do?",
    "Which payment methods do you accept?"
]

for q in queries:
    print(f"Customer: {q}")
    print(f"Bot:      {support_chain.invoke(q)}\n")
````

### Use Case 2 — Code Review Assistant

**Problem**: Dev team wants automated reviews catching bugs and security vulnerabilities.

**Components**: PromptTemplate → HuggingFaceEndpoint → StrOutputParser

````python
from langchain_huggingface import HuggingFaceEndpoint
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=300,
    temperature=0,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

review_prompt = PromptTemplate.from_template(
    """You are a senior Python developer. Review the code below.
Identify: 1) Security vulnerabilities 2) Bugs 3) Improvements.

Code:
{code}

Review:"""
)

review_chain = review_prompt | llm | StrOutputParser()

# Intentionally buggy + insecure code
bad_code = """
def get_user(user_id):
    query = "SELECT * FROM users WHERE id = " + user_id  # SQL injection!
    return db.execute(query)

def calculate_average(numbers):
    total = 0
    for n in numbers:
        total = total + n
    return total / len(numbers)  # ZeroDivisionError if list is empty!
"""

print("Code Review:")
print(review_chain.invoke({"code": bad_code}))
````

### Use Case 3 — Research Assistant with Web Search

**Problem**: Analyst needs current-event answers backed by live web search results.

**Components**: DuckDuckGoSearchRun (free, no API key) + HuggingFace LLM + LCEL Chain

````python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_huggingface import HuggingFaceEndpoint
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# DuckDuckGo — completely free, no API key needed
search = DuckDuckGoSearchRun()

llm = HuggingFaceEndpoint(
    repo_id="google/flan-t5-large",
    task="text2text-generation",
    max_new_tokens=200,
    temperature=0,
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"]
)

research_prompt = PromptTemplate.from_template(
    """Answer the question based on the search results below.
Be accurate and concise.

Search results:
{search_results}

Question: {question}
Answer:"""
)

research_chain = research_prompt | llm | StrOutputParser()

def research(question: str) -> str:
    """Search the web and answer using LLM."""
    print(f"Searching for: {question}")
    results = search.run(question)
    print(f"Search preview: {results[:250]}...\n")
    return research_chain.invoke({"search_results": results, "question": question})

answer = research("What is LangGraph and how is it different from LangChain?")
print("Answer:", answer)
````

## 6. Advantages and Limitations

### Strengths

| Strength | Why it matters |
|---|---|
| **Modularity** | Swap HuggingFace → OpenAI → Ollama with one line |
| **Rapid prototyping** | Working RAG chatbot in under 50 lines |
| **Ecosystem** | 100+ loaders, 50+ vector stores, dozens of LLM providers |
| **LCEL** | Chains are composable, streamable, parallelizable |
| **Provider agnostic** | Works with free models and paid APIs equally |

### Limitations

| Limitation | Impact |
|---|---|
| **Abstraction leakage** | Hard to debug failures inside a chain |
| **Fast-changing API** | v0.1 → v0.2 → v0.3 deprecated many patterns |
| **Agent latency** | Multiple LLM calls = higher latency and token cost |
| **Overkill for simple tasks** | Direct API call is faster for single-turn queries |

### When NOT to use LangChain
- Single-turn summarization or classification (use raw API)
- Strict sub-100ms latency SLAs (agent loops can't meet this)
- When just starting to learn LLMs — understand the raw API first

---
## 7. Conclusion

### Key Takeaways
- **LCEL** (`|` pipe operator) is the modern way to build chains — learn it first
- **RAG** is the most important real-world pattern — most enterprise LLM apps are RAG apps
- **Agents** are powerful but expensive — use only when static chains aren't sufficient
- **HuggingFace** gives you free access to capable models for learning and prototyping
- **sentence-transformers** gives you high-quality local embeddings with zero cost

### Future Scope
- **LangGraph** — graph-based stateful workflows for multi-actor agents (loops, branches, parallel)
- **Multi-agent systems** — Researcher + Coder + Reviewer agents under a Supervisor
- **LangSmith** — observability, tracing, and evaluation for production LangChain apps

LangChain moves you from *calling* an LLM to *designing* full AI systems — a core skill for real-world GenAI engineering.
