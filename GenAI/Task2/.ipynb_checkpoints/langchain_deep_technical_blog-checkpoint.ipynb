## 1. Introduction to LangChain

**LangChain** is an open-source Python framework for building applications powered by Large Language Models (LLMs).

It solves 3 core problems raw LLMs have:

| Problem | LangChain Solution |
|---|---|
| No memory across turns | Memory modules |
| Can't access external data | Document Loaders + Vector Stores (RAG) |
| Can't take actions | Agents + Tools |
| Brittle prompt strings | Prompt Templates |
| Hard to chain LLM calls | Chains (LCEL) |

---

## 2. Core Components

### 2.1 LLMs and Chat Models

LangChain wraps all LLM providers behind a unified interface. Swap OpenAI for HuggingFace by changing one line — your chain logic stays unchanged.

````python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

# Initialize the model
llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7
)

# Basic LLM call
messages = [
    SystemMessage(content="You are a concise technical assistant."),
    HumanMessage(content="What is a transformer model in one sentence?")
]

response = llm.invoke(messages)
print("Response:", response.content)
````

### 2.2 Prompts and Prompt Templates

A `PromptTemplate` separates prompt structure from content — making prompts reusable, testable, and version-controllable.

````python
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate

# Simple string template with variables
basic_template = PromptTemplate.from_template(
    "Explain {concept} to a {audience} in {num_sentences} sentences."
)

formatted = basic_template.format(
    concept="recursion",
    audience="10-year-old",
    num_sentences="2"
)
print("Formatted prompt:", formatted)

# Chat prompt template for multi-turn models
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in {domain}. Be precise."),
    ("human", "Question: {question}")
])

messages = chat_template.format_messages(
    domain="machine learning",
    question="What is gradient descent?"
)
print("\nChat messages:")
for msg in messages:
    print(f"  [{msg.type}]: {msg.content}")
````

### 2.3 Chains (LCEL — LangChain Expression Language)

Chains connect prompt → LLM → output parser using the `|` pipe operator. This is the modern LCEL approach introduced in LangChain 0.2+.

````python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
parser = StrOutputParser()

# Define prompt template
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a Python expert. Provide only code, no explanation."),
    ("human", "Write a Python function that {task}")
])

# Build chain using the pipe | operator (LCEL)
chain = prompt | llm | parser

# Invoke the chain
result = chain.invoke({"task": "reverses a list without using the reverse() method"})
print("Generated code:")
print(result)
````

### 2.4 Memory

Memory modules inject conversation history back into the prompt, giving a stateless LLM the ability to "remember" previous turns.

````python
from langchain_openai import ChatOpenAI
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.5)

# Buffer memory — stores full conversation history
memory = ConversationBufferMemory(return_messages=True)

# ConversationChain automatically injects history into every prompt
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=False
)

# Turn 1
response1 = conversation.predict(input="My name is Rahul. I'm learning LangChain.")
print("Turn 1:", response1)

# Turn 2 — the model remembers context from Turn 1
response2 = conversation.predict(input="What is my name and what am I learning?")
print("\nTurn 2:", response2)

# Inspect raw memory contents
print("\nMemory stored:", len(memory.chat_memory.messages), "messages")
````

### 2.5 Agents

An Agent uses an LLM as a reasoning engine to decide which tools to call and in what order — dynamically, based on the query. It follows the **ReAct** loop: Think → Act → Observe → repeat.

````python
from langchain_openai import ChatOpenAI
from langchain.tools import tool
from langchain.agents import AgentExecutor, create_openai_tools_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

# Define a custom tool using the @tool decorator
@tool
def multiply(a: float, b: float) -> float:
    """Multiplies two numbers. Use this for any multiplication task."""
    return a * b

@tool
def get_string_length(text: str) -> int:
    """Returns the character length of a string."""
    return len(text)

tools = [multiply, get_string_length]

# Agent prompt — must include agent_scratchpad for reasoning
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant. Use tools when needed."),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

# Create agent and executor
agent = create_openai_tools_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

# Run the agent
result = agent_executor.invoke({"input": "What is 45.5 multiplied by 8.2?"})
print("\nFinal Answer:", result["output"])
````

### 2.6 Tools

Tools are functions exposed to agents. They have a `name`, `description` (LLM uses this to decide when to call them), and an `args_schema`.

````python
from langchain.tools import tool, StructuredTool
from pydantic import BaseModel

# Decorator-based tool
@tool
def get_word_count(text: str) -> int:
    """Returns the number of words in a given text."""
    return len(text.split())

# Structured tool with Pydantic schema
class WeatherInput(BaseModel):
    city: str
    unit: str = "celsius"

def fetch_weather(city: str, unit: str = "celsius") -> str:
    """Mock weather function — replace with real API call in production."""
    return f"The temperature in {city} is 28°{unit[0].upper()}."

weather_tool = StructuredTool.from_function(
    func=fetch_weather,
    name="get_weather",
    description="Fetches current weather for a city. Use when user asks about weather.",
    args_schema=WeatherInput
)

# Inspect tool properties
print("Tool name:", get_word_count.name)
print("Tool description:", get_word_count.description)
print("Direct call result:", get_word_count.run("Hello world from LangChain"))

print("\nWeather tool name:", weather_tool.name)
print("Weather tool result:", weather_tool.run({"city": "Hyderabad", "unit": "celsius"}))
````

### 2.7 Document Loaders

Document Loaders ingest data from PDFs, websites, CSVs, and dozens of other sources, returning a standard `Document` object.

````python
from langchain.schema import Document
from langchain_community.document_loaders import WebBaseLoader

# Load a live webpage
loader = WebBaseLoader("https://python.langchain.com/docs/introduction/")
docs = loader.load()

print(f"Documents loaded: {len(docs)}")
print(f"Source: {docs[0].metadata.get('source', 'N/A')}")
print(f"Content preview (first 300 chars):\n{docs[0].page_content[:300]}")

# --- Simulating a PDF/text load without a file ---
# In real usage: PyPDFLoader('your.pdf').load()
sample_docs = [
    Document(page_content="LangChain was created by Harrison Chase in October 2022.", metadata={"source": "wiki"}),
    Document(page_content="LangChain supports Python and JavaScript runtimes.", metadata={"source": "wiki"}),
    Document(page_content="LCEL uses the pipe operator | for composing chains.", metadata={"source": "docs"}),
]
print(f"\nSample documents created: {len(sample_docs)}")
print(f"First doc metadata: {sample_docs[0].metadata}")
````

### 2.8 Indexes and Vector Stores

Text chunks are embedded into vectors and stored in a vector database. At query time, semantic similarity search retrieves the most relevant chunks — enabling RAG (Retrieval-Augmented Generation).

````python
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document

# Sample knowledge base
raw_texts = [
    "LangChain was created by Harrison Chase in October 2022.",
    "LangChain supports both Python and JavaScript.",
    "LCEL (LangChain Expression Language) uses the pipe | operator for chaining.",
    "Agents in LangChain use the ReAct (Reasoning + Acting) framework.",
    "FAISS is a vector store library developed by Meta AI Research.",
    "LangSmith is the observability and evaluation platform for LangChain apps.",
    "LangGraph is used for building multi-actor, stateful agentic workflows."
]

# Create Document objects
docs = [Document(page_content=t, metadata={"id": i}) for i, t in enumerate(raw_texts)]

# Split into chunks (important for large documents)
splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=20)
chunks = splitter.split_documents(docs)
print(f"Total chunks after splitting: {len(chunks)}")

# Create embeddings and FAISS index
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
vector_store = FAISS.from_documents(chunks, embeddings)

# Semantic similarity search
query = "Who built LangChain and when?"
results = vector_store.similarity_search(query, k=2)

print(f"\nQuery: '{query}'")
print("Top results:")
for i, doc in enumerate(results):
    print(f"  [{i+1}] {doc.page_content}")
````

## 3. Architecture Diagram

```
User Input
    │
    ▼
Prompt Template  ← injects system context, few-shot examples
    │
    ▼
LLM / Chat Model  ← OpenAI, HuggingFace, Anthropic, Ollama
    │
    ▼
Chain             ← sequences calls, routes output
   ╱ ╲
  ╱   ╲
Tools  Memory
(APIs) (history)
  ╲   ╱
   ╲ ╱
  Output

Document Loaders → Vector Stores → Retriever feeds into Chain as context (RAG)
```

**Flow explanation:**
1. User query enters via the input interface
2. PromptTemplate formats it with system instructions and examples
3. LLM processes the formatted prompt
4. Chain orchestrates the call and routes output
5. If dynamic action is needed → Agent picks a Tool
6. Memory injects conversation history for context
7. Final answer returned to the user

## 4. Hands-on: Full RAG Pipeline

````python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain.schema import Document

# --- Knowledge base ---
texts = [
    "LangChain was created by Harrison Chase in October 2022.",
    "LangChain supports Python and JavaScript.",
    "LCEL uses the pipe operator | for composing chains.",
    "Agents use the ReAct (Reasoning + Acting) framework.",
    "FAISS is a vector store by Meta AI. It supports billion-scale similarity search.",
    "LangSmith provides observability, tracing, and evaluation for LangChain apps.",
    "LangGraph enables stateful, multi-actor agentic workflows using graph-based state machines."
]

# --- Build vector store ---
docs = [Document(page_content=t) for t in texts]
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
vector_store = FAISS.from_documents(docs, embeddings)
retriever = vector_store.as_retriever(search_kwargs={"k": 2})

# --- RAG prompt ---
rag_prompt = ChatPromptTemplate.from_messages([
    ("system", """Answer using ONLY the context below.
If the answer isn't there, say 'I don't know.'

Context:
{context}"""),
    ("human", "{question}")
])

# --- RAG chain ---
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

def format_docs(docs):
    """Join retrieved document chunks into a single context string."""
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | rag_prompt
    | llm
    | StrOutputParser()
)

# --- Test queries ---
questions = [
    "Who created LangChain and when?",
    "What is LCEL?",
    "What does LangSmith do?"
]

for q in questions:
    answer = rag_chain.invoke(q)
    print(f"Q: {q}")
    print(f"A: {answer}\n")
````

## 5. Real-World Use Cases

### Use Case 1: Customer Support Chatbot

**Problem**: E-commerce company needs a chatbot that answers questions about orders, returns, and policies — remembering earlier messages in the conversation.

**Components**: DocumentLoader → FAISS → ConversationSummaryMemory → RAG Chain

````python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.memory import ConversationBufferMemory
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from langchain.schema import Document

# Mock company knowledge base
faq_docs = [
    Document(page_content="Refund policy: Customers can return items within 30 days for a full refund."),
    Document(page_content="Shipping: Standard delivery takes 3-5 business days. Express is 1-2 days."),
    Document(page_content="To track your order, visit our website and enter your order ID in the tracking portal."),
    Document(page_content="Damaged items: Contact support within 48 hours with a photo. Replacement shipped in 24 hours."),
]

# Build retriever
embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
vs = FAISS.from_documents(faq_docs, embeddings)
retriever = vs.as_retriever(search_kwargs={"k": 2})

# Support bot chain
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)

support_prompt = ChatPromptTemplate.from_messages([
    ("system", """You are a friendly customer support agent.
Use the FAQ context to answer. Be concise and helpful.

FAQ Context:
{context}"""),
    ("human", "{question}")
])

def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)

support_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | support_prompt
    | llm
    | StrOutputParser()
)

# Test the support bot
queries = [
    "What is your return policy?",
    "My item arrived damaged, what do I do?"
]

for q in queries:
    print(f"Customer: {q}")
    print(f"Bot: {support_chain.invoke(q)}\n")
````

### Use Case 2: Code Review Assistant

**Problem**: Dev team wants automated LLM-powered code reviews that flag bugs, security issues, and style violations.

**Components**: PromptTemplate → ChatOpenAI → StrOutputParser

````python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)

review_prompt = ChatPromptTemplate.from_messages([
    ("system", """You are a senior software engineer reviewing Python code.
Check for:
1. Bugs and logic errors
2. Security vulnerabilities (especially injection attacks)
3. PEP8 style violations
4. Performance improvements

Format your response with clear numbered sections."""),
    ("human", "Review this Python code:\n\n```python\n{code}\n```")
])

review_chain = review_prompt | llm | StrOutputParser()

# Intentionally buggy and insecure code for review
bad_code = """
def get_user(user_id):
    # Fetch user from database
    query = "SELECT * FROM users WHERE id = " + user_id
    return db.execute(query)

def calculate_average(numbers):
    total = 0
    for n in numbers:
        total = total + n
    return total / len(numbers)  # Will crash on empty list!
"""

review = review_chain.invoke({"code": bad_code})
print(review)
````

### Use Case 3: Research Assistant with Web Search

**Problem**: Analyst needs LLM answers about recent events (beyond training cutoff), with search-backed sources.

**Components**: ChatOpenAI + DuckDuckGoSearchRun + AgentExecutor

````python
from langchain_openai import ChatOpenAI
from langchain_community.tools import DuckDuckGoSearchRun
from langchain.agents import AgentExecutor, create_openai_tools_agent
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

llm = ChatOpenAI(model="gpt-4o-mini", temperature=0)
search = DuckDuckGoSearchRun()
tools = [search]

research_prompt = ChatPromptTemplate.from_messages([
    ("system", """You are a research assistant. Use web search for current events.
Always cite your sources in the answer."""),
    ("human", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])

agent = create_openai_tools_agent(llm, tools, research_prompt)
executor = AgentExecutor(agent=agent, tools=tools, verbose=True, max_iterations=3)

result = executor.invoke({"input": "What are the latest LangChain updates in 2025?"})
print("\n--- Final Answer ---")
print(result["output"])
````

## 6. Advantages and Limitations

### ✅ Strengths

| Strength | Why it matters |
|---|---|
| **Modularity** | Swap OpenAI → Ollama by changing one line |
| **Rapid prototyping** | Working RAG chatbot in <50 lines |
| **Ecosystem breadth** | 100+ loaders, 50+ vector stores, dozens of LLM providers |
| **LCEL** | Chains are composable, streamable, parallelizable |

### ❌ Limitations

| Limitation | Impact |
|---|---|
| **Abstraction leakage** | Hard to debug failures deep in a chain |
| **Rapidly changing API** | v0.1 → v0.2 → v0.3 broke code frequently |
| **Agent latency** | Multiple LLM calls = high latency + cost |
| **Overkill for simple tasks** | Direct API call is faster for single-turn tasks |

### 🚫 When NOT to use LangChain
- Single-turn summarization or classification (use raw API)
- Strict latency SLAs where agent loops are unacceptable
- Learning LLMs for the first time (understand raw API first)

---
## 7. Conclusion

### Key Takeaways
- **LCEL** is the modern way to build chains — learn it first
- **RAG** is the most practically important pattern for enterprise apps
- **Agents** are powerful but expensive — use only when static chains aren't sufficient
- Always monitor token usage — agent loops can get expensive fast

### Future Scope
- **LangGraph**: Graph-based state machines for multi-actor agentic workflows
- **Multi-agent systems**: Researcher + Coder + Reviewer agents collaborating
- **LangSmith**: Observability, tracing, and LLM evaluation in production

LangChain moves you from *using* an LLM to *designing* full AI systems — a critical skill for real-world GenAI engineering roles.
