# Problem 1: Social Media – Post Engagement Analyzer
## Problem Statement
Create a function to analyze engagement on a social media post.

Rules
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement

Real-Time Use
-
- Social media analytics platforms
- Marketing dashboards



```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes = total_likes + likes
    
    # Check if post is viral
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 150, 300, 400]  
analyze_engagement(likes)
```

# Problem 2: Healthcare – Medicine Stock Alert System
## Problem Statement
Create a function to monitor medicine stock levels.
    
Rules
- Stock < 10 → Low Stock Alert
- Stock ≥ 10 → Stock Sufficient

Real-Time Use
- Pharmacy management systems
- Hospital inventory tracking


```python
def medicine_stock_alert(stock):
    if stock >= 10:
        status = "Stock Sufficient"
    else:
        status = "Low Stock Alert"
        
    print("Medicine Stock: ", stock)
    print("Status: ",status)
    
stock = int(input("Enter the number of stocks: "))
medicine_stock_alert(stock)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.

Rules
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

Real-Time Use
-
- Climate monitoring systems
- Irrigation planning tools


```python
# Function to check rainfall adequacy
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    # Loop to calculate rainfall
    for rain in rainfall_data:
        total = total + rain
        
    #average rainfall
    avg = total/len(rainfall_data)
    
    if avg > required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
        
    #output
    print("Average Rainfall: ", int(avg))
    print("Rainfall Status: ", status)


# Example usage
season_rainfall = [70, 75, 68, 80, 67]
required = 70

rainfall_checker(season_rainfall, required)
```
