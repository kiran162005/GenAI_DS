# Problem 1: Social Media – Post Engagement Analyzer
## Problem Statement
Create a function to analyze engagement on a social media post.

Rules
- Calculate total likes
- If Total Likes ≥ 1000 → Viral Post
- Otherwise → Normal Engagement

Real-Time Use
-
- Social media analytics platforms
- Marketing dashboards



```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for likes in likes_list:
        total_likes = total_likes + likes
    
    # Check if post is viral
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    # Display output
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 150, 300, 400]  
analyze_engagement(likes)
```

# Problem 2: Healthcare – Medicine Stock Alert System
## Problem Statement
Create a function to monitor medicine stock levels.
    
Rules
- Stock < 10 → Low Stock Alert
- Stock ≥ 10 → Stock Sufficient

Real-Time Use
- Pharmacy management systems
- Hospital inventory tracking


```python
# Function to monitor medicine stock
def medicine_stock_alert(stock):
    
    # Conditional logic to check stock level
    if stock >= 10:
        status = "Stock Sufficient"
    else:
        status = "Low Stock Alert"

    #Display output
    print("Medicine Stock: ", stock)
    print("Status: ",status)

#Example usage
stock = int(input("Enter the number of stocks: "))
medicine_stock_alert(stock)
```

# Problem 3: Agriculture – Rainfall Adequacy Checker
## Problem Statement
Create a function that analyzes seasonal rainfall data.

Rules
- Calculate average rainfall
- Average ≥ required level → Adequate Rainfall
- Otherwise → Inadequate Rainfall

Real-Time Use
-
- Climate monitoring systems
- Irrigation planning tools


```python
# Function to check rainfall adequacy
def rainfall_checker(rainfall_data, required_level):
    total = 0
    
    # Loop to calculate rainfall
    for rain in rainfall_data:
        total = total + rain
        
    #average rainfall
    avg = total/len(rainfall_data)
    
    if avg > required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
        
    #output
    print("Average Rainfall: ", int(avg))
    print("Rainfall Status: ", status)


# Example usage
season_rainfall = [70, 75, 68, 80, 67]
required = 70

rainfall_checker(season_rainfall, required)
```

# Problem 4: Social Media – Duplicate Account Detection
## Problem Statement
Create a function to detect duplicate usernames in a system.

Rules
- If duplicates exist → Flag accounts
- Else → Accounts are unique

Real-Time Use
- Platform moderation
- Fraud detection systems



```python
# Function to detect duplicate usernames
def duplicate_account_detection(usernames):
    unique_users = set(usernames)   # Convert list to set

    # Check if duplicates exist
    if len(unique_users) < len(usernames):
        print("Duplicate accounts found: Yes")
    else:
        print("Duplicate accounts found: No")

# Example usage
accounts = ["user1", "user2", "user3", "user1", "admin"]
duplicate_account_detection(accounts)
```

## Problem 6: Agriculture – Premium Crop Price Filter
## Problem Statement
Create a function that filters crop prices above a premium threshold.

Rules
- Price > ₹2000 → Premium Crop
- Else → Regular Crop

Real-Time Use
- Market analytics
- Export pricing systems


```python
# Function to filter premium crop prices
def premium_crops_filter(prices):
    
    #Empty list to store the prices
    premium_crops = []
    
    #  Loop to filter prices
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    #Display output
    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crops_filter(crop_prices)
    
```
