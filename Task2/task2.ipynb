## Problem 1: Unique Words in a Sentence
# Problem Statement

Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

# Input
Enter sentence: python is easy and python is powerful

# Expected Output

Unique words count: 5  
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}

## Real-Time Applications

- Text analytics

- Natural Language Processing (NLP) preprocessing

- Search engine indexing

- Chat analysis systems


```python
# Unique Words in a Sentence

# Input sentence
sentence = "python is easy and python is powerful"

# Split sentence into words
words = sentence.split()

# Convert to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

# Problem 2: Highest Salary from Employee Data
## Problem Statement
Store employee names and their salaries, and determine which employee has the highest salary.

## Input Format
    
A dictionary where:

Keys represent employee names

Values represent employee salaries

Example Input

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

## Expected Output

Highest Salary: Ravi - 75000

## Real-Time Applications
- HR management systems
- Payroll analysis


```python
# Highest Salary from Employee Data

# Dictionary of employees and salaries
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Assume first employee as highest initially
highest_employee = list(employees.keys())[0]
highest_salary = employees[highest_employee]

# Loop through dictionary
for name in employees:
    if employees[name] > highest_salary:
        highest_salary = employees[name]
        highest_employee = name

print("Highest Salary:", highest_employee, "-", highest_salary)
```

```python
#Problem 3: Find Maximum and Minimum Values

# List of numbers
numbers = [45, 22, 89, 10, 66]

# Assume first element as max and min
maximum = numbers[0]
minimum = numbers[0]

# Loop through numbers
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
#Problem 4: Count Products Above a Price Threshold

# List of product prices
prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count = count + 1

print("Products above 1000:", count)
```

```python
# Problem 5: Calculate Attendance Percentage

# Attendance list
attendance = ["P", "P", "A", "P", "P"]

present = 0
total_days = len(attendance)

# Count number of "P"
for day in attendance:
    if day == "P":
        present = present + 1

percentage = (present / total_days) * 100

print("Attendance Percentage:", percentage)
```

```python
# Problem 7: Count Character Frequency

# Input string
text = "python"

# Empty dictionary
frequency = {}

# Loop through each character
for char in text:
    if char in frequency:
        frequency[char] = frequency[char] + 1
    else:
        frequency[char] = 1

print(frequency)
```
