# Problem Statement 1:Smart Parking Lot Management System
## Design a function to manage a smart parking lot.
The system should:
- Accept vehicle entry and exit logs
- Calculate total parked vehicles
- Identify peak parking usage
- Alert if parking exceeds capacity


**Real-Time Use**
- Mall parking systems
- Smart city infrastructure


```python
def smart_parking_system(capacity, vehical):
    current_parked = 0
    peak_usage = 0


    #loop through vehical logs
    for log in vehical:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            if current_parked > 0:
                current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    #Parking status check
    if current_parked > capacity:
        status = "Alert! Parking Exceeded Capacity"
    elif current_parked == capacity:
        status = "Parking Full"
    else:
        status = "Availabe"

    # Display results
        print("Currently Parked Vehicles:", current_parked)
        print("Peak Parking Usage:", peak_usage)
        print("Parking Status:", status)

# Sample Input
capacity = 50
vehicle = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle)
```

# Problem Statement 2:Online Food Delivery Time Estimator
## Create a function that estimates delivery time based on:
- Distance (km)
- Weather condition
- Traffic level

Apply delays dynamically and display final ETA.

**Real-Time Use**
- Food delivery apps
- Logistics platforms


```python
def estimate_delivery_time(distance, traffic, weather):
    
    # Base travel time (2 minutes per km)
    base_time = distance * 2
    
    # Fixed preparation time
    prep_time = 20
    
    # Traffic delay
    if traffic == "Low":
        traffic_delay = 5
    elif traffic == "Medium":
        traffic_delay = 10
    elif traffic == "High":
        traffic_delay = 15
    else:
        traffic_delay = 0
    
    # Weather delay
    if weather == "Clear":
        weather_delay = 0
    elif weather == "Cloudy":
        weather_delay = 2
    elif weather == "Rainy":
        weather_delay = 4
    elif weather == "Storm":
        weather_delay = 8
    else:
        weather_delay = 0
    
    # Final ETA
    total_time = base_time + prep_time + traffic_delay + weather_delay
    
    print("Estimated Delivery Time:", total_time, "minutes")

# Sample input
estimate_delivery_time(8, "High", "Rainy")
```

# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
## Build a function that analyzes seat booking data and:
- Calculates occupancy percentage
- Determines if show is Housefull
- Suggests opening additional shows

**Real-Time Use**
- Cinema ticketing systems
- Event management software



```python
def movie_occupancy_analyzer(total_seats, booked_seats):
    
    # Count total booked seats
    booked_count = len(booked_seats)
    
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    # Determine show status
    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    elif occupancy >= 50:
        status = "Moderate Booking"
    else:
        status = "Seats Available"
    
    # Suggest additional show
    if occupancy >= 90:
        suggestion = "Consider Opening Additional Show"
    else:
        suggestion = "No Additional Show Needed"
    
    # Display results
    print("Occupancy:", int(occupancy), "%")
    print("Show Status:", status)
    print("Suggestion:", suggestion)

# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

movie_occupancy_analyzer(total_seats, booked_seats)
```

# Problem Statement 4:Cloud Server Load Classification System
## Create a function to classify server load based on CPU usage readings.
**Rules:**
- Average CPU < 50% → Normal
- 50%–80% → Warning
- 80% → Critical

**Real-Time Use**
- Cloud monitoring dashboards
- DevOps alerting systems


```python
def classify_server_load(cpu_readings):
    
    total = 0
    
    # Loop to calculate total CPU usage
    for reading in cpu_readings:
        total += reading
    
    # Calculate average
    average = total / len(cpu_readings)
    
    # Classification
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    
    # Display results
    print("Average CPU Load:", int(average), "%")
    print("Server Status:", status)

# Sample input
cpu_data = [45, 60, 70, 85, 90]
classify_server_load(cpu_data)
```
