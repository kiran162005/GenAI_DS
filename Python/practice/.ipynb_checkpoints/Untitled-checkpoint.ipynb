```python
#hotel booking system

def calculate_hotel_bill(age,days,room_type):
    base_price = 3000
    if age< 18:
        return "Booking rejected: Age must be 18 or above"
    if days <= 0 or days > 10:
        return "Booking rejected: Invalid Stay days"
    if room_type not in ["standard", "deluxe"]:
        return "Booking Rejected:Invalid room type"
    total = base_price * days
    if room_type == "deluxe":
        total += 2000*days
    return f"Booking Sucessful! Total Bill is : {total}"

#user input
age = int(input("Enter age: "))
days = int(input("Enter days: "))
room_type = input("Enter room type(standard/deluxe):").lower()

#call function
result = calculate_hotel_bill(age,days,room_type)
print(result)
```

```python
#Loan eligibility app

def loan_eligibility(age,salary,credit_score):
    if age < 21 or age > 60:
        return "Loan rejected"
    if salary < 25000:
        return "Loan Rejected: Salary is less than 25000"
    if credit_score < 700:
        return "Loan rejected"
    return "Loan Approved"

age = int(input("Enter age: "))
salary = int(input("Enter Salary: "))
credit_score = int(input("Enter score"))

#call
result = loan_eligibility(age,salary,credit_score)
print(result)
```

```python
#Function to classify patients heart rate and risk level
def classify_heart_rate(heart_rate, activity_status):
    if activity_status == 'rest':
        if heart_rate < 60:
            status = "Low"
        elif heart_rate <= 100:
            status = "Normal"
        else:
            status = "High"
            
        #check of patient is at active
    elif activity_status == "Active":
        if heart_rate < 60:
            status = "Low"
        elif heart_rate <= 120:
            status = "Normal"
        else:
            status = "High"
    else:
        return "Invalid activity status"
    #risk alert rules
    if "Normal" in status:
        risk = "Stable condition"
    else:
        risk = "Medical attention required"
        
        
    #return final result
    return status,risk

#user inputs
heart_rate = int(input("Enter patient's heart rate: "))
activity_status = input("Enter activity stastus(rest/active):").lower()

result = classify_heart_rate(heart_rate,activity_status)
print(result)


```

# Crop yield analysis

```python
def analyze_yield(yields):
    total = 0

#calculate total yield
    for value in yields:
        total = total + value
    #calculate avg
    average = total/len(yields)
    #decide yield status
    if average >= 50:
        status = "Good production"
    else:
        status = "Bad position"
    return total,average,status

#empty list
daily_yields = []
#ask days
days = int(input("Enter the number of days: "))
for i in range(days):
    yield_value = int(input(f"Enter yield for day{i+1}(kg): "))
    daily_yields.append(yield_value)

#call function
total, average, result = analyze_yield(daily_yields)
#output
print("Daily yields: ",daily_yields)
print("Total: ",total)
print("Average: ",average)
print("Status: ",result)

```

```python
def calculate_fare(stations,peak):
    base_fare = 20
    fare = base_fare + (stations * 5)
    if peak == 'yes':
        fare += fare * 0.20
    if fare > 100:
        fare = 100
    return fare

#user inputs
stations = int(input("Enter no. of stations travelled: "))
peak = input("is it peak hour?(yes/no): ").lower()

#Function call
final_fare = calculate_fare(stations,peak)
print(final_fare)
```
